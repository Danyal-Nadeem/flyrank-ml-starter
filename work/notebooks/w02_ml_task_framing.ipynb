## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
Scoring — I'm assigning each page a priority score for content refresh,
rather than a binary yes/no classification. Scoring fits because "should
this be refreshed" is a matter of degree (how stale + how visible + how
much CTR is being lost), not a clean category.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
Target (proxy): is_declining_label, derived from trend_direction == "down".
This is an observed outcome from the data, not a human-defined rule — it
reflects what actually happened to the page's performance, which I then
combine with impressions_90d to prioritize which declining pages matter most.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
Precision@50 — of the top 50 pages my score flags for refresh, what
fraction are genuinely declining pages. "Good" means beating the hand-rule
baseline (0.680 from notebook 02) since that's the honest bar to clear.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
A fixed rule like "stale AND visible" treats every matching page as equally
urgent. In reality, a page that's slightly stale with a sharp CTR drop can
matter more than a very stale page with flat metrics. ML can weigh multiple
signals together and rank by combined strength, catching priority cases a
single if-statement threshold would miss or misrank.
```
